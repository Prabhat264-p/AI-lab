```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

digits = load_digits()

print("Dataset Shape:", digits.data.shape)
print("Image Shape:", digits.images.shape)
print("Target Labels:", digits.target_names)

plt.figure(figsize=(4, 4))
plt.imshow(digits.images[0], cmap='pink')
plt.title(f"Label: {digits.target[0]}")
plt.axis('off')
plt.show()
```

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

digits = load_digits()

plt.figure(figsize=(10, 15))

shown = set()
plot_index = 1

for i in range(len(digits.target)):
    label = digits.target[i]

    if label not in shown and label <= 9:
        plt.subplot(5, 2, plot_index)
        plt.imshow(digits.images[i], cmap='Reds')
        plt.title(f" {label}")
        plt.axis('off')

        shown.add(label)
        plot_index += 1

plt.tight_layout()
plt.show()
```
